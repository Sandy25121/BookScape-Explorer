```python
import pandas as pd
import sqlalchemy  as sa
from sqlalchemy import create_engine
from sqlalchemy import text
```

```python
sql_con = sa.engine.URL.create(
    drivername = 'mysql+pymysql',
    username = "root",
    password = "mysql",
    host = "127.0.0.1",
    database ="bookscape"
)

sql_engine = create_engine(sql_con)
```

```python
# 1 . Check Availability of eBooks vs Physical Books

query = '''SELECT isEbook,
        COUNT(*) AS book_count 
        FROM books 
        GROUP BY isEbook;'''

df = pd.read_sql_query(query,sql_engine)
df
```

```python
#2. Find the Publisher with the Most Books Published
query = ''' SELECT publisher, 
        Count(*) as book_count
        FROM BOOKS
        Group by publisher
        Order By book_count DESC
        Limit 5 offset 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
#3. Identify the Publisher with the Highest Average Rating

query = ''' SELECT publisher, averageRating, count(averageRating) as rating_count
        FROM BOOKS
        group by publisher, averageRating
        Order By rating_count DESC
        Limit 1 offset 1
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
#4. Get the Top 5 Most Expensive Books by Retail Price

query = ''' SELECT book_title, amount_retailPrice, currencyCode_retailPrice
        FROM BOOKS
        Order By amount_retailPrice DESC
        Limit 5
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
#5. Find Books Published After 2010 with at Least 500 Pages

query = ''' SELECT book_title, year, pageCount
        FROM BOOKS where year >= 2010 and pageCount >=500
        order by year 
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 6. List Books with Discounts Greater than 20%

query = ''' SELECT book_title, amount_listPrice, amount_retailPrice, 
       (amount_listPrice - amount_retailPrice) / amount_listPrice * 100 AS discount_percentage
        FROM books
        WHERE amount_listPrice > 0 
        AND (amount_listPrice - amount_retailPrice) / amount_listPrice * 100 > 20;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 7. Find the Average Page Count for eBooks vs Physical Books
query = ''' SELECT isEbook, AVG(pageCount) AS avg_page_count
        FROM books
        GROUP BY isEbook;
         '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 8. Find the Top 3 Authors with the Most Books

query = ''' SELECT book_authors, COUNT(*) AS book_count
        FROM books
        GROUP BY book_authors
        ORDER BY book_count DESC
        LIMIT 3 offset 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 9. List Publishers with More than 10 Books

query = ''' SELECT publisher, COUNT(*) AS book_count
        FROM books
        GROUP BY publisher
        HAVING book_count > 10;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 10. Find the Average Page Count for Each Category

query = ''' SELECT categories, AVG(pageCount) AS avg_page_count
        FROM books
        GROUP BY categories;
       '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 11. Retrieve Books with More than 3 Authors

query = ''' SELECT book_title, book_authors
        FROM books
        WHERE LENGTH(book_authors) - LENGTH(REPLACE(book_authors, ',', '')) + 1 > 3;
       '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 12. Books with Ratings Count Greater Than the Average


query = ''' SELECT book_title, ratingsCount
        FROM books
        WHERE ratingsCount > (SELECT AVG(ratingsCount) FROM books);
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 13. Books with the Same Author Published in the Same Year

query = ''' SELECT book_authors, year, COUNT(*) AS book_count
        FROM books
        GROUP BY book_authors, year
        HAVING book_count > 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 14. Books with a Specific Keyword in the Title

query = '''SELECT book_title
           FROM books
           WHERE book_title LIKE '%%Python%%';
        '''
df = pd.read_sql_query(query, sql_engine)
df
```

```python
# 15. Year with the Highest Average Book Price

query = ''' SELECT year, AVG(amount_retailPrice) AS avg_price
            FROM books
            GROUP BY year
            ORDER BY avg_price DESC
            LIMIT 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 16. Count Authors Who Published 3 Consecutive Years

query = ''' SELECT book_authors, COUNT(DISTINCT year) AS consecutive_years
        FROM books
        GROUP BY book_authors
        HAVING MAX(year) - MIN(year) >= 2;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 17. Authors Who Published in the Same Year Under Different Publishers

query = ''' SELECT book_authors, year, COUNT(DISTINCT publisher) AS publisher_count
        FROM books
        GROUP BY book_authors, year
        HAVING publisher_count > 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 18. Average Retail Price of eBooks vs Physical Books

query = '''SELECT 
        AVG(CASE WHEN isEbook = 1 THEN amount_retailPrice END) AS avg_ebook_price,
        AVG(CASE WHEN isEbook = 0 THEN amount_retailPrice END) AS avg_physical_price
        FROM books;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
# 19. Identify Books with title, averageRating, and ratingsCount for these outliers.

query = '''SELECT book_title, averageRating, ratingsCount
FROM books
WHERE averageRating > (
        SELECT AVG(averageRating) + 2 * STDDEV(averageRating)
        FROM books
    )
OR averageRating < (
        SELECT AVG(averageRating) - 2 * STDDEV(averageRating)
        FROM books
    );
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```

```python
#20. Publisher with the Highest Average Rating

query = '''SELECT publisher, AVG(averageRating) AS avg_rating, COUNT(*) AS book_count
          FROM books
          GROUP BY publisher
          HAVING book_count > 10
          ORDER BY avg_rating DESC
          LIMIT 1;
        '''
        
df = pd.read_sql_query(query,sql_engine)
df
```
